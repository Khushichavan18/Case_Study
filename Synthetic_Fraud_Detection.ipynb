```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("synthetic_fraud_dataset.csv")


sns.set_style("whitegrid")
sns.set_context("paper")  


plt.figure(figsize=(8, 5))
plt.scatter(df['Transaction_Amount'], df['Risk_Score'], alpha=0.5, c=df['Fraud_Label'], cmap='coolwarm')


plt.xlabel("Transaction Amount")
plt.ylabel("Risk Score")
plt.title("Transaction Amount vs Risk Score (Fraud Detection)")
plt.colorbar(label="Fraud Label (0 = No, 1 = Yes)") 


plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Transaction_Amount", y="Risk_Score", hue="Fraud_Label", data=df, palette="coolwarm", alpha=0.7)
plt.title("Transaction Amount vs Risk Score")
plt.xlabel("Transaction Amount")
plt.ylabel("Risk Score")
plt.legend(title="Fraud Label")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(8, 5))

# Create histogram
plt.hist(df["Transaction_Amount"], bins=30, color="blue", alpha=0.7, edgecolor="black")

# Labels and title
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.title("Distribution of Transaction Amounts")

# Show plot
plt.show()
```

```python
import seaborn as sns

# Set Seaborn style
sns.set_style("darkgrid")

# Create histogram
plt.figure(figsize=(8, 5))
sns.histplot(df["Transaction_Amount"], bins=30, kde=True, color="blue")

# Labels and title
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.title("Transaction Amount Distribution")

# Show plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df[df["Fraud_Label"] == 1]["Transaction_Amount"], bins=30, kde=True, color="red", label="Fraud", alpha=0.6)
sns.histplot(df[df["Fraud_Label"] == 0]["Transaction_Amount"], bins=30, kde=True, color="blue", label="Non-Fraud", alpha=0.6)

# Labels and title
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.title("Transaction Amount Distribution (Fraud vs Non-Fraud)")
plt.legend()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set_style("whitegrid")

# Create bar plot
plt.figure(figsize=(6, 4))
sns.countplot(x="Fraud_Label", data=df, palette="coolwarm")

# Labels and title
plt.xlabel("Fraud (0 = No, 1 = Yes)")
plt.ylabel("Transaction Count")
plt.title("Fraud vs Non-Fraud Transactions")

# Show plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="Transaction_Type", hue="Fraud_Label", data=df, palette="coolwarm")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Labels and title
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.title("Fraud Count Across Transaction Types")
plt.legend(title="Fraud Label", labels=["Non-Fraud", "Fraud"])

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (Replace with actual file path if needed)
df = pd.read_csv("synthetic_fraud_dataset.csv")  # Update with the actual dataset file

# Convert timestamp to datetime if necessary
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")

# Drop non-numeric columns for correlation heatmap
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title and show plot
plt.title("Correlation Heatmap of Fraud Detection Dataset")
plt.show()
```
